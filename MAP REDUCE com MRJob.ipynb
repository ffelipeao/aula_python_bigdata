## Introdução ao MapReduce

**MapReduce** é um modelo de programação e um sistema para processamento distribuído de grandes conjuntos de dados (Big Data) em clusters de computadores. Ele foi originalmente desenvolvido pelo Google para indexar a web e se tornou a base para muitas ferramentas de processamento de Big Data, como o Hadoop.

### Como Funciona?

O processo MapReduce é dividido em duas fases principais:

1.  **Fase Map (Mapeamento):**
    *   A função `Map` pega um conjunto de pares chave-valor de entrada.
    *   Para cada par, ela processa os dados e gera um conjunto de pares chave-valor intermediários. Em essência, ela "mapeia" os dados brutos para um formato mais estruturado ou relevante para a análise.

2.  **Fase Reduce (Redução):**
    *   A função `Reduce` pega os pares chave-valor intermediários gerados pela fase `Map`.
    *   Ela agrupa todos os valores associados a uma mesma chave intermediária e então aplica uma operação de "redução" (como soma, contagem, média, etc.) para combinar esses valores e produzir um conjunto menor de resultados finais.

### Exemplo: Contagem de Palavras

Vamos usar o exemplo de contagem de palavras para ilustrar o MapReduce:

**Dados de Entrada:**
"Isso é um exemplo de texto. Este texto será processado pelo MRJob. Exemplo de MRJob."

**1. Fase Map:**
O *mapper* leria cada linha do texto e, para cada palavra encontrada, emitiria um par `(palavra, 1)`. Por exemplo:

*   ("Isso", 1)
*   ("é", 1)
*   ("um", 1)
*   ("exemplo", 1)
*   ("de", 1)
*   ("texto", 1)
*   ("Este", 1)
*   ("texto", 1)
*   ("será", 1)
*   ("processado", 1)
*   ("pelo", 1)
*   ("MRJob", 1)
*   ("Exemplo", 1)
*   ("de", 1)
*   ("MRJob", 1)

**2. Embaralhamento e Classificação (Shuffle & Sort):**
Os pares intermediários são agrupados por chave. Isso significa que todas as ocorrências da mesma palavra são coletadas juntas. Exemplo:

*   ("de", [1, 1])
*   ("é", [1])
*   ("exemplo", [1, 1])
*   ("isso", [1])
*   ("mrjob", [1, 1])
*   ("pelo", [1])
*   ("processado", [1])
*   ("será", [1])
*   ("este", [1])
*   ("texto", [1, 1])
*   ("um", [1])

**3. Fase Reduce:**
O *reducer* recebe cada chave com sua lista de valores e soma os valores para obter a contagem final da palavra. Exemplo:

*   ("de", 2)
*   ("é", 1)
*   ("exemplo", 2)
*   ("isso", 1)
*   ("mrjob", 2)
*   ("pelo", 1)
*   ("processado", 1)
*   ("será", 1)
*   ("este", 1)
*   ("texto", 2)
*   ("um", 1)

### Utilidade do MapReduce

*   **Processamento de Big Data:** É ideal para conjuntos de dados que são grandes demais para caber na memória de uma única máquina.
*   **Paralelização:** O modelo é inerentemente paralelo, permitindo que as tarefas sejam divididas e executadas simultaneamente em múltiplos nós de um cluster, acelerando o processamento.
*   **Tolerância a Falhas:** Se um nó falhar durante o processamento, o MapReduce pode redistribuir a tarefa para outro nó e continuar o trabalho, garantindo a robustez do sistema.
*   **Escalabilidade:** A capacidade de adicionar mais máquinas ao cluster para lidar com volumes de dados crescentes o torna altamente escalável.
*   **Variedade de Casos de Uso:** Além da contagem de palavras, é usado para indexação de busca, análise de logs, processamento de grafos, machine learning e muito mais.

```python
# Instalar (se necessário)
#!pip install mrjob

# Criar o script Python REAL
%%writefile job.py
from mrjob.job import MRJob
import re
import logging

log = logging.getLogger(__name__)

palavra_regex = re.compile(r"[\w']+")

class QuantidadePalavras(MRJob):

    def mapper(self, _, linha):
        log.info("Executando MAPPER")
        for p in palavra_regex.findall(linha):
            yield (p.lower(), 1)

    def reducer(self, p, qtd):
        log.info("Executando REDUCER")
        yield (p, sum(qtd))

if __name__ == '__main__':
    QuantidadePalavras.run()
```
