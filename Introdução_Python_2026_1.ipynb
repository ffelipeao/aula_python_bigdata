Diferenças entre tupla, lista e dicionário em Python, com exemplos:

1. Lista (list)
Mutável (pode ser alterada após a criação).
Permite elementos repetidos.
Usa colchetes [].


````python
lista = [1, 2, 3, 4, 4]
print(lista)
lista.append(5)  # Adiciona um elemento
print(lista)  # [1, 2, 3, 4, 5]
````

2. Tupla (tuple)
Imutável (não pode ser alterada após a criação).
Usa parênteses ().
Útil para armazenar dados fixos.

````python
tupla = (1, 2, 3, 4, 5)
# tupla.append(5)
# tupla[0] = 10  # Isso daria erro, pois tuplas são imutáveis
print(tupla)  # (1, 2, 3, 4)
````

3. Dicionário (dict)
Mutável.
Armazena pares chave-valor.
Usa chaves {}.

````python
dicionario = {"nome": "Ana", "idade": 25, "cidade": "São Paulo"}
print(dicionario["nome"])  # Ana
dicionario["idade"] = 26  # Alterando um valor
dicionario["cidade"] = 'Rio de Janeiro'
print(dicionario)
````

💡 Resumo Rápido:

Lista → Ordenada, mutável, usa [].
Tupla → Ordenada, imutável, usa ().
Dicionário → Não ordenado (até Python 3.6), mutável, usa {} com chave-valor.

List Comprehension

  List comprehension é uma construção sintática do Python que permite criar uma nova lista a partir de outra lista. É uma forma rápida e eficiente de construir listas, que simplifica o código e reduz a necessidade de loops explícitos.

Como funciona:

  Para cada valor de uma lista, aplica-se uma operação e coloca-se todos os resultados em uma nova lista Combina a iteração e a expressão de seleção em uma única linha Filtra e transforma elementos de uma sequência com facilidade


Vantagens:

  Torna o código mais curto e fácil de ler
  Geralmente, as listas de compreensão são mais rápidas do que loops tradicionais
  Menos linhas de código significam menos espaço para erros e mais clareza na lógica

Estrutura:

  Expression: A expressão que define o que cada elemento da nova lista será
  For: Um loop for que itera sobre cada item em um iterável
  Item in iterable: O iterável sobre o qual o loop for é executado

````python
quadrados = [x**2 for x in range(10)]
print(quadrados)
````

## Dicionário de Dicionários

Um dicionário de dicionários (ou dicionário aninhado) é uma estrutura de dados onde os valores de um dicionário são, por sua vez, outros dicionários. Isso permite representar dados com múltiplos níveis de hierarquia.

### Como criar:

Você pode criar um dicionário de dicionários definindo dicionários internos como valores para as chaves de um dicionário externo. Cada dicionário interno pode ter suas próprias chaves e valores.

### Exemplo:

Imagine que você quer armazenar informações sobre diferentes cidades, e para cada cidade, você quer guardar dados como população e atrações principais. Você pode usar um dicionário de dicionários para isso.

````python
cidades = {
    "São Paulo": {
        "populacao": 12000000,
        "estado": "SP",
        "atracoes": ["Parque Ibirapuera", "Museu de Arte de São Paulo"]
    },
    "Rio de Janeiro": {
        "populacao": 6700000,
        "estado": "RJ",
        "atracoes": ["Cristo Redentor", "Pão de Açúcar"]
    },
    "Belo Horizonte": {
        "populacao": 2500000,
        "estado": "MG",
        "atracoes": ["Lagoa da Pampulha", "Mercado Central"]
    }
}
````

````python
print(cidades["Belo Horizonte"]["atracoes"][1])
````

````python
for cidade, dados_cidade in cidades.items():
  print(cidade)
  for chave, valor in dados_cidade.items():
    print('-----', chave, valor)
````

````python
# Acessando informações
print("População de São Paulo:", cidades["São Paulo"]["populacao"])
print("Estado do Rio de Janeiro:", cidades["Rio de Janeiro"]["estado"])
print("Atrações de Belo Horizonte:", cidades["Belo Horizonte"]["atracoes"])

# Iterando sobre o dicionário de dicionários
print("\n--- Informações das Cidades ---")
for cidade, dados_cidade in cidades.items():
    print(f"\nCidade: {cidade}")
    for chave_dado, valor_dado in dados_cidade.items():
         print(f"  {chave_dado.capitalize()}: {valor_dado}")
````

## Manipulação de Arquivos com `with open()`

`with open()` é a forma **recomendada** em Python para abrir e manipular arquivos (leitura, escrita, etc.). Sua principal vantagem é garantir que o arquivo seja **sempre fechado automaticamente**, mesmo que ocorram erros durante a operação.

### Por que usar `with open()`?

*   **Segurança**: Evita vazamento de recursos, pois o arquivo é fechado automaticamente ao sair do bloco `with`.
*   **Conveniência**: Não é necessário chamar `arquivo.close()` explicitamente.
*   **Legibilidade**: Torna o código mais limpo e fácil de entender.

### Como usar?

A sintaxe básica é:

```python
with open('nome_do_arquivo.txt', 'modo') as variavel_do_arquivo:
    # Faça suas operações de leitura/escrita aqui
```

**Modos comuns:**

*   `'r'` (read): Para ler o conteúdo de um arquivo (padrão).
*   `'w'` (write): Para escrever em um arquivo. **Cria um novo arquivo se não existir ou sobrescreve o existente.**
*   `'a'` (append): Para adicionar conteúdo ao final de um arquivo. Cria um novo arquivo se não existir.

### Exemplos:

**1. Escrever em um arquivo:**

```python
with open('meu_arquivo.txt', 'w') as arquivo:
    arquivo.write('Olá, mundo!\n')
    arquivo.write('Esta é a segunda linha.')
```

**2. Ler de um arquivo:**

```python
with open('meu_arquivo.txt', 'r') as arquivo:
    conteudo = arquivo.read()
    print(conteudo)
```

````python
with open("arquivo.txt", "w") as arquivo:
    arquivo.write("\n Tiago")
    arquivo.write("\n Ana")
    arquivo.write("\n José")
    arquivo.write("\n Mateus")
````

## Classes e Orientação a Objetos (POO)

A **Programação Orientada a Objetos (POO)** é um paradigma de programação que organiza o software em torno de **objetos**, em vez de apenas funções e dados. O conceito central da POO é a **Classe**.

### O que é uma Classe?

Uma **classe** é como um molde ou um "projeto" para criar objetos. Ela define as características (chamadas de **atributos**) e os comportamentos (chamados de **métodos** ou funções) que os objetos criados a partir dela terão. Pense em uma classe `Carro` que define que todo carro tem uma cor, um modelo e pode acelerar ou frear.

### O que é um Objeto?

Um **objeto** é uma instância concreta de uma classe. Se a classe `Carro` é o projeto, então seu carro específico, com sua cor e modelo, é um objeto da classe `Carro`. Você pode ter vários objetos (vários carros) criados a partir da mesma classe, cada um com seus próprios valores para os atributos.

### Princípios Chave da POO (brevemente):

1.  **Encapsulamento**: Agrupar dados (atributos) e métodos que operam nesses dados em uma única unidade (o objeto), escondendo os detalhes internos da implementação.
2.  **Herança**: Permite que uma nova classe (subclasse ou classe filha) herde atributos e métodos de uma classe existente (superclasse ou classe pai), promovendo a reutilização de código.
3.  **Polimorfismo**: Capacidade de objetos de diferentes classes de responder ao mesmo método de maneiras diferentes.

### Exemplo Simples em Python:

```python
class Cachorro:
    # Atributo da classe
    especie = "Canis familiaris"

    # Método construtor: chamado ao criar um novo objeto
    def __init__(self, nome, idade):
        self.nome = nome  # Atributo de instância
        self.idade = idade # Atributo de instância

    # Método do objeto
    def latir(self):
        return f"{self.nome} está latindo: Au Au!"

# Criando objetos (instâncias da classe Cachorro)
meu_cachorro = Cachorro("Bob", 5)
outro_cachorro = Cachorro("Rex", 2)

# Acessando atributos dos objetos
print(f"Nome do meu cachorro: {meu_cachorro.nome}") # Saída: Nome do meu cachorro: Bob
print(f"Idade do outro cachorro: {outro_cachorro.idade}") # Saída: Idade do outro cachorro: Rex

# Chamando métodos dos objetos
print(meu_cachorro.latir())  # Saída: Bob está latindo: Au Au!
print(outro_cachorro.latir()) # Saída: Rex está latindo: Au Au!

# Acessando atributo da classe
print(f"Espécie: {Cachorro.especie}") # Saída: Espécie: Canis familiaris
```

````python
#Criando minha classe cachorro
class Cachorro:
    # Atributo da classe
    especie = "Canis familiaris"

    # Método construtor: chamado ao criar um novo objeto
    def __init__(self, nome, idade, cor):
        self.nome = nome  # Atributo de instância
        self.idade = idade # Atributo de instância
        self.cor=cor

    # Método do objeto
    def latir(self, latido):
        return f"{self.nome} está latindo: {latido}!"

    def andar(self):
        return f"{self.nome} está andando"

    def dormir(self):
        return f"{self.nome} está dormindo"

    def brincar(self):
        return f"{self.nome} está brincando"
````

````python
cachorro1 = Cachorro("Bob", 5, 'Caramelo')
cachorro2 = Cachorro("Rex", 2, 'Marrom')
print(cachorro1.dormir())
print(cachorro2.brincar())

print(cachorro1.latir('Au au aua uauauauauauauauauauau'))
print(cachorro2.cor)
````

### Mais Exemplos com Pandas (DataFrames e Análise Básica)

O Pandas é uma biblioteca poderosa para manipulação e preparação de dados em Python. Com funções similares às encontradas em Excel ou SQL, o Pandas facilita a ordenação, filtragem, agrupamento e agregação de dados.

Usada para manipulação e análise de dados tabulares em Python, usando as estruturas de dados `DataFrame` e `Series`.

**Exemplo 1: Criação e Seleção de Dados em um DataFrame**

````python
import pandas as pd

# Criando um DataFrame a partir de um dicionário
dados = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Idade': [24, 27, 22, 32, 29],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'São Paulo', 'Curitiba'],
    'Pontuação': [85, 92, 78, 95, 88]
}
df = pd.DataFrame(dados)

print("DataFrame Original:")
display(df)

# Selecionando uma única coluna (resulta em uma Series)
nomes = df['Nome']
print("\nColuna 'Nome':")
print(nomes)

# Selecionando múltiplas colunas (resulta em um DataFrame)
cidades_idades = df[['Cidade', 'Idade']]
print("\nColunas 'Cidade' e 'Idade':")
display(cidades_idades)

# Filtrando linhas com base em uma condição
alunos_maiores_25 = df[df['Idade'] > 25]
print("\nAlunos com mais de 25 anos:")
display(alunos_maiores_25)

alunos_sao_paulo = df[df['Cidade'] == 'São Paulo']
print("\nAlunos de São Paulo:")
display(alunos_sao_paulo)
````

**Exemplo 2: Estatísticas Descritivas e Agrupamento**

Pandas oferece métodos fáceis para sumarizar e agrupar dados.

````python
import pandas as pd

dados = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Idade': [24, 27, 22, 32, 29],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'São Paulo', 'Curitiba'],
    'Pontuação': [85, 92, 78, 95, 88]
}
df = pd.DataFrame(dados)

# Estatísticas descritivas para colunas numéricas
estatisticas = df.describe()
print("\nEstatísticas Descritivas do DataFrame:")
display(estatisticas)

# Média da pontuação
media_pontuacao = df['Pontuação'].mean()
print("\nMédia da pontuação:", media_pontuacao)

# Agrupando dados por 'Cidade' e calculando a média da 'Pontuação'
media_por_cidade = df.groupby('Cidade')['Pontuação'].mean()
print("\nMédia da pontuação por Cidade:")
print(media_por_cidade)

# Contagem de valores únicos em uma coluna
contagem_cidades = df['Cidade'].value_counts()
print("\nContagem de alunos por Cidade:")
print(contagem_cidades)
````

### Mais Exemplos com Matplotlib (Visualização de Dados)

O Matplotlib é uma biblioteca do Python para visualização de dados, permitindo criar diversos tipos de gráficos, **como linhas, barras, dispersão (scatter), histogramas e muitos outros.** Ele é altamente personalizável e amplamente utilizado em ciência de dados, engenharia e análise estatística.

É a base para muitas outras bibliotecas de visualização.

**Exemplo 1: Gráfico de Linha Simples**

Um gráfico de linha é ideal para mostrar tendências ao longo do tempo ou de uma sequência ordenada.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
x = np.linspace(0, 10, 100) # 100 pontos entre 0 e 10
y1 = np.sin(x)
y2 = np.cos(x)

# Criando o gráfico
plt.figure(figsize=(10, 6)) # Define o tamanho da figura (largura, altura)
plt.plot(x, y1, label='Função Seno', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Função Cosseno', color='red', linestyle='--', linewidth=2)

# Adicionando título e rótulos aos eixos
plt.title('Gráfico de Funções Seno e Cosseno', fontsize=16)
plt.xlabel('Eixo X', fontsize=12)
plt.ylabel('Eixo Y', fontsize=12)

# Adicionando uma legenda
plt.legend(fontsize=10)

# Adicionando grade
plt.grid(True, linestyle=':', alpha=0.7)

# Personalizando os ticks dos eixos
plt.xticks(np.arange(0, 11, 1)) # Ticks de 1 em 1 no eixo X
plt.yticks(np.arange(-1, 1.1, 0.5)) # Ticks de 0.5 em 0.5 no eixo Y

# Exibindo o gráfico
plt.show()
````

**Exemplo 2: Gráfico de Barras e Dispersão (Scatter Plot)**

*   **Gráfico de Barras**: Ótimo para comparar quantidades entre diferentes categorias.
*   **Gráfico de Dispersão**: Excelente para visualizar a relação entre duas variáveis numéricas.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Dados para gráfico de barras
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 12, 39]

plt.figure(figsize=(14, 6))

# Subplot 1: Gráfico de Barras
plt.subplot(1, 2, 1) # 1 linha, 2 colunas, primeiro gráfico
plt.bar(categorias, valores, color=['skyblue', 'lightcoral', 'lightgreen', 'orange', 'purple'])
plt.title('Valores por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Dados para gráfico de dispersão
pessoas = {
    'Idade': [25, 30, 35, 40, 28, 32, 38, 27, 42, 33],
    'Renda': [3000, 4500, 5000, 6500, 3200, 4800, 5500, 3100, 7000, 4900]
}
df_pessoas = pd.DataFrame(pessoas)

# Subplot 2: Gráfico de Dispersão
plt.subplot(1, 2, 2) # 1 linha, 2 colunas, segundo gráfico
plt.scatter(df_pessoas['Idade'], df_pessoas['Renda'], color='green', alpha=0.7, s=100) # s é o tamanho dos pontos
plt.title('Relação entre Idade e Renda')
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout() # Ajusta o layout para evitar sobreposição
plt.show()
````

### Mais Exemplos com `requests` (Requisições HTTP)

A biblioteca `requests` simplifica a interação com APIs web e a realização de requisições HTTP (GET, POST, PUT, DELETE, etc.).

**Exemplo 1: Fazendo uma Requisição GET e Processando JSON**

Vamos usar a API pública `JSONPlaceholder`, que oferece dados de teste (posts, comentários, etc.).

````python
import requests
import json

# URL da API para obter uma lista de posts
url = 'https://jsonplaceholder.typicode.com/posts/1' # Obtém o post com ID 1

print(f"Fazendo requisição GET para: {url}")

try:
    # Realiza a requisição GET
    response = requests.get(url)

    # Verifica o status da requisição
    # Um código 200 indica sucesso
    print(f"Status Code: {response.status_code}")

    # Se a requisição foi bem sucedida (status 200)
    if response.status_code == 200:
        # O conteúdo da resposta é geralmente JSON. Podemos parseá-lo diretamente
        dados_post = response.json()

        print("\nDados do Post (JSON):")
        # Imprime o JSON de forma formatada para melhor legibilidade
        print(json.dumps(dados_post, indent=2))

        print(f"\nTítulo do Post: {dados_post['title']}")
        print(f"Corpo do Post: {dados_post['body'][:50]}...") # Primeiros 50 caracteres

    elif response.status_code == 404:
        print("Erro: Recurso não encontrado (404 Not Found).")
    else:
        print(f"Erro na requisição: {response.status_code} - {response.text}")

except requests.exceptions.RequestException as e:
    print(f"Ocorreu um erro de conexão: {e}")
````

**Exemplo 2: Enviando Dados com uma Requisição POST**

Requisições POST são usadas para enviar dados para um servidor, geralmente para criar um novo recurso. Vamos simular a criação de um novo post.

````python
import requests
import json

# URL da API para criar um novo post
url = 'https://jsonplaceholder.typicode.com/posts'

# Dados que queremos enviar no corpo da requisição
novo_post = {
    'title': 'Meu Novo Post Teste',
    'body': 'Este é o conteúdo do meu post enviado via Python requests.',
    'userId': 1
}

print(f"Fazendo requisição POST para: {url}")
print("Dados a serem enviados:")
print(json.dumps(novo_post, indent=2))

try:
    # Realiza a requisição POST
    # Os dados são enviados como JSON, então usamos `json=` e definimos o header `Content-Type`
    response = requests.post(url, json=novo_post)

    # Verifica o status da requisição
    # Para criação, um código 201 (Created) é o esperado
    print(f"\nStatus Code: {response.status_code}")

    if response.status_code == 201:
        # O servidor geralmente retorna o recurso criado, incluindo seu novo ID
        dados_resposta = response.json()

        print("\nResposta do Servidor (Novo Post Criado):")
        print(json.dumps(dados_resposta, indent=2))

        print(f"\nNovo Post criado com ID: {dados_resposta.get('id')}")
    else:
        print(f"Erro ao criar post: {response.status_code} - {response.text}")

except requests.exceptions.RequestException as e:
    print(f"Ocorreu um erro de conexão: {e}")
````

**Geradores**

O que são Geradores (Generators) em Python?

Geradores são uma forma especial de criar iteradores em Python de maneira simples e eficiente.

Em vez de retornar todos os valores de uma vez (como uma lista), um gerador produz os valores sob demanda, um de cada vez, usando a palavra-chave yield.

👉 Isso economiza memória e é muito útil quando lidamos com muitos dados.

📌 Diferença básica: return vs yield

*   return → encerra a função e devolve um valor.

*   yield → pausa a função e devolve um valor, mas mantém o estado da execução.

Quando a função for chamada novamente, ela continua exatamente de onde parou.

````python
def contador(x):
    for i in range(x):
        yield i
print(list(contador(5)))
````

Expressões Regulares

O código utiliza a biblioteca re (expressões regulares) para encontrar números em uma string.

Explicação:
re.search(r"\d+", "A123B")

O padrão \d+ busca um ou mais dígitos (\d = número, + = uma ou mais ocorrências).
A string "A123B" contém "123", que corresponde ao padrão.
padrao.group()

Retorna a parte da string que corresponde ao padrão encontrado, que neste caso é "123".

````python
import re
padrao = re.search(r"\d+", "A123B")
print(padrao.group())  # Saída: 123
````

### Exemplo de Leitura de JSON Externo via HTTP

Vamos usar a biblioteca `requests` para fazer uma requisição HTTP GET para uma URL externa que retorna dados em formato JSON, e em seguida, processar esses dados.

````python
import requests
import json

# URL do serviço JSON externo
url_json_externo = 'https://my-json-server.typicode.com/maujor/livros-json/db'

print(f"Fazendo requisição GET para: {url_json_externo}")

try:
    # Realiza a requisição GET
    response = requests.get(url_json_externo)

    # Verifica o status da requisição
    print(f"Status Code: {response.status_code}")

    # Se a requisição foi bem sucedida (status 200)
    if response.status_code == 200:
        # O conteúdo da resposta é JSON. Podemos parseá-lo diretamente
        dados_json = response.json()

        print("\nDados do JSON Externo (Primeiras linhas):")
        # Imprime o JSON de forma formatada para melhor legibilidade
        # E mostra apenas uma parte para não poluir a saída
        print(json.dumps(dados_json, indent=2))

        # Exemplo de como acessar dados específicos
        if 'livros' in dados_json and len(dados_json['livros']) > 0:
            print(f"\nTotal de livros encontrados: {len(dados_json['livros'])}")
            print(f"Primeiro livro: {dados_json['livros'][0]['titulo']} de {dados_json['livros'][0]['autor']}")
        else:
            print("Nenhum livro encontrado ou formato inesperado.")

    else:
        print(f"Erro na requisição: {response.status_code} - {response.text}")

except requests.exceptions.RequestException as e:
    print(f"Ocorreu um erro de conexão: {e}")
````

Exemplo de como podemos usar a IA para facilitar nosso trabalho.

Usando a ia para traduzir os dados durante o processo de leitura de uma arquivo CSV.


````python
import pandas as pd
from google.colab import ai

# Caminho para o arquivo CSV de exemplo
file_path = '/content/sample_data/california_housing_test.csv'

# 1. Carregar o CSV em um DataFrame
df_original = pd.read_csv(file_path)

print("### DataFrame Original (Primeiras 5 linhas):")
display(df_original.head())
````

````python
# 2. Obter os nomes das colunas originais
original_column_names = df_original.columns.tolist()

# 3. Traduzir cada nome de coluna usando google.colab.ai
translated_column_names = []
print("\n### Traduzindo nomes das colunas...")
for col_name in original_column_names:
    prompt = f"Traduza o seguinte nome de coluna em inglês para português: '{col_name}'. Retorne apenas a tradução." # Adicionado instrução para retornar apenas a tradução
    translated_text = ai.generate_text(prompt)
    # Realiza uma limpeza simples para remover aspas ou formatação extra da resposta da IA
    cleaned_translation = translated_text.strip().replace("'", "").replace('"', '')
    translated_column_names.append(cleaned_translation)
    print(f"Original: '{col_name}' -> Traduzido: '{cleaned_translation}'")

# 4. Criar um novo DataFrame com os nomes das colunas traduzidos
df_translated = df_original.copy()
df_translated.columns = translated_column_names

print("\n### DataFrame com Colunas Traduzidas (Primeiras 5 linhas):")
display(df_translated.head())
````
