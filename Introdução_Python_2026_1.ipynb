Diferenças entre tupla, lista e dicionário em Python, com exemplos:

1. Lista (list)
Mutável (pode ser alterada após a criação).
Permite elementos repetidos.
Usa colchetes [].


````python
lista = [1, 2, 3, 4, 4]
print(lista)
lista.append(5)  # Adiciona um elemento
print(lista)  # [1, 2, 3, 4, 5]
````

2. Tupla (tuple)
Imutável (não pode ser alterada após a criação).
Usa parênteses ().
Útil para armazenar dados fixos.

````python
tupla = (1, 2, 3, 4, 5)
# tupla.append(5)
# tupla[0] = 10  # Isso daria erro, pois tuplas são imutáveis
print(tupla)  # (1, 2, 3, 4)
````

3. Dicionário (dict)
Mutável.
Armazena pares chave-valor.
Usa chaves {}.

````python
dicionario = {"nome": "Ana", "idade": 25, "cidade": "São Paulo"}
print(dicionario["nome"])  # Ana
dicionario["idade"] = 26  # Alterando um valor
dicionario["cidade"] = 'Rio de Janeiro'
print(dicionario)
````

💡 Resumo Rápido:

Lista → Ordenada, mutável, usa [].
Tupla → Ordenada, imutável, usa ().
Dicionário → Não ordenado (até Python 3.6), mutável, usa {} com chave-valor.

List Comprehension

  List comprehension é uma construção sintática do Python que permite criar uma nova lista a partir de outra lista. É uma forma rápida e eficiente de construir listas, que simplifica o código e reduz a necessidade de loops explícitos.

Como funciona:

  Para cada valor de uma lista, aplica-se uma operação e coloca-se todos os resultados em uma nova lista Combina a iteração e a expressão de seleção em uma única linha Filtra e transforma elementos de uma sequência com facilidade


Vantagens:

  Torna o código mais curto e fácil de ler
  Geralmente, as listas de compreensão são mais rápidas do que loops tradicionais
  Menos linhas de código significam menos espaço para erros e mais clareza na lógica

Estrutura:

  Expression: A expressão que define o que cada elemento da nova lista será
  For: Um loop for que itera sobre cada item em um iterável
  Item in iterable: O iterável sobre o qual o loop for é executado

````python
quadrados = [x**2 for x in range(10)]
print(quadrados)
````

## Dicionário de Dicionários

Um dicionário de dicionários (ou dicionário aninhado) é uma estrutura de dados onde os valores de um dicionário são, por sua vez, outros dicionários. Isso permite representar dados com múltiplos níveis de hierarquia.

### Como criar:

Você pode criar um dicionário de dicionários definindo dicionários internos como valores para as chaves de um dicionário externo. Cada dicionário interno pode ter suas próprias chaves e valores.

### Exemplo:

Imagine que você quer armazenar informações sobre diferentes cidades, e para cada cidade, você quer guardar dados como população e atrações principais. Você pode usar um dicionário de dicionários para isso.

````python
cidades = {
    "São Paulo": {
        "populacao": 12000000,
        "estado": "SP",
        "atracoes": ["Parque Ibirapuera", "Museu de Arte de São Paulo"]
    },
    "Rio de Janeiro": {
        "populacao": 6700000,
        "estado": "RJ",
        "atracoes": ["Cristo Redentor", "Pão de Açúcar"]
    },
    "Belo Horizonte": {
        "populacao": 2500000,
        "estado": "MG",
        "atracoes": ["Lagoa da Pampulha", "Mercado Central"]
    }
}

# Acessando informações
print("População de São Paulo:", cidades["São Paulo"]["populacao"])
print("Estado do Rio de Janeiro:", cidades["Rio de Janeiro"]["estado"])
print("Atrações de Belo Horizonte:", cidades["Belo Horizonte"]["atracoes"])

# Iterando sobre o dicionário de dicionários
print("\n--- Informações das Cidades ---")
for cidade, dados_cidade in cidades.items():
    print(f"\nCidade: {cidade}")
    for chave_dado, valor_dado in dados_cidade.items():
         print(f"  {chave_dado.capitalize()}: {valor_dado}")
````

## Manipulação de Arquivos com `with open()`

`with open()` é a forma **recomendada** em Python para abrir e manipular arquivos (leitura, escrita, etc.). Sua principal vantagem é garantir que o arquivo seja **sempre fechado automaticamente**, mesmo que ocorram erros durante a operação.

### Por que usar `with open()`?

*   **Segurança**: Evita vazamento de recursos, pois o arquivo é fechado automaticamente ao sair do bloco `with`.
*   **Conveniência**: Não é necessário chamar `arquivo.close()` explicitamente.
*   **Legibilidade**: Torna o código mais limpo e fácil de entender.

### Como usar?

A sintaxe básica é:

```python
with open('nome_do_arquivo.txt', 'modo') as variavel_do_arquivo:
    # Faça suas operações de leitura/escrita aqui
```

**Modos comuns:**

*   `'r'` (read): Para ler o conteúdo de um arquivo (padrão).
*   `'w'` (write): Para escrever em um arquivo. **Cria um novo arquivo se não existir ou sobrescreve o existente.**
*   `'a'` (append): Para adicionar conteúdo ao final de um arquivo. Cria um novo arquivo se não existir.

### Exemplos:

**1. Escrever em um arquivo:**

```python
with open('meu_arquivo.txt', 'w') as arquivo:
    arquivo.write('Olá, mundo!\n')
    arquivo.write('Esta é a segunda linha.')
```

**2. Ler de um arquivo:**

```python
with open('meu_arquivo.txt', 'r') as arquivo:
    conteudo = arquivo.read()
    print(conteudo)
```

````python
with open("arquivo.txt", "w") as arquivo:
    arquivo.write("\n Tiago")
    arquivo.write("\n Ana")
    arquivo.write("\n José")
    arquivo.write("\n Mateus")
````

## Classes e Orientação a Objetos (POO)

A **Programação Orientada a Objetos (POO)** é um paradigma de programação que organiza o software em torno de **objetos**, em vez de apenas funções e dados. O conceito central da POO é a **Classe**.

### O que é uma Classe?

Uma **classe** é como um molde ou um "projeto" para criar objetos. Ela define as características (chamadas de **atributos**) e os comportamentos (chamados de **métodos** ou funções) que os objetos criados a partir dela terão. Pense em uma classe `Carro` que define que todo carro tem uma cor, um modelo e pode acelerar ou frear.

### O que é um Objeto?

Um **objeto** é uma instância concreta de uma classe. Se a classe `Carro` é o projeto, então seu carro específico, com sua cor e modelo, é um objeto da classe `Carro`. Você pode ter vários objetos (vários carros) criados a partir da mesma classe, cada um com seus próprios valores para os atributos.

### Princípios Chave da POO (brevemente):

1.  **Encapsulamento**: Agrupar dados (atributos) e métodos que operam nesses dados em uma única unidade (o objeto), escondendo os detalhes internos da implementação.
2.  **Herança**: Permite que uma nova classe (subclasse ou classe filha) herde atributos e métodos de uma classe existente (superclasse ou classe pai), promovendo a reutilização de código.
3.  **Polimorfismo**: Capacidade de objetos de diferentes classes de responder ao mesmo método de maneiras diferentes.

### Exemplo Simples em Python:

```python
class Cachorro:
    # Atributo da classe
    especie = "Canis familiaris"

    # Método construtor: chamado ao criar um novo objeto
    def __init__(self, nome, idade):
        self.nome = nome  # Atributo de instância
        self.idade = idade # Atributo de instância

    # Método do objeto
    def latir(self):
        return f"{self.nome} está latindo: Au Au!"

# Criando objetos (instâncias da classe Cachorro)
meu_cachorro = Cachorro("Bob", 5)
outro_cachorro = Cachorro("Rex", 2)

# Acessando atributos dos objetos
print(f"Nome do meu cachorro: {meu_cachorro.nome}") # Saída: Nome do meu cachorro: Bob
print(f"Idade do outro cachorro: {outro_cachorro.idade}") # Saída: Idade do outro cachorro: Rex

# Chamando métodos dos objetos
print(meu_cachorro.latir())  # Saída: Bob está latindo: Au Au!
print(outro_cachorro.latir()) # Saída: Rex está latindo: Au Au!

# Acessando atributo da classe
print(f"Espécie: {Cachorro.especie}") # Saída: Espécie: Canis familiaris
```

**Geradores**

O que são Geradores (Generators) em Python?

Geradores são uma forma especial de criar iteradores em Python de maneira simples e eficiente.

Em vez de retornar todos os valores de uma vez (como uma lista), um gerador produz os valores sob demanda, um de cada vez, usando a palavra-chave yield.

👉 Isso economiza memória e é muito útil quando lidamos com muitos dados.

📌 Diferença básica: return vs yield

*   return → encerra a função e devolve um valor.

*   yield → pausa a função e devolve um valor, mas mantém o estado da execução.

Quando a função for chamada novamente, ela continua exatamente de onde parou.

````python
def contador(x):
    for i in range(x):
        yield i
print(list(contador(5)))
````

Expressões Regulares

O código utiliza a biblioteca re (expressões regulares) para encontrar números em uma string.

Explicação:
re.search(r"\d+", "A123B")

O padrão \d+ busca um ou mais dígitos (\d = número, + = uma ou mais ocorrências).
A string "A123B" contém "123", que corresponde ao padrão.
padrao.group()

Retorna a parte da string que corresponde ao padrão encontrado, que neste caso é "123".

````python
import re
padrao = re.search(r"\d+", "A123B")
print(padrao.group())  # Saída: 123
````

Exemplo de como podemos usar a IA para facilitar nosso trabalho.

Usando a ia para traduzir os dados durante o processo de leitura de uma arquivo CSV.


````python
import pandas as pd
from google.colab import ai

# Caminho para o arquivo CSV de exemplo
file_path = '/content/sample_data/california_housing_test.csv'

# 1. Carregar o CSV em um DataFrame
df_original = pd.read_csv(file_path)

print("### DataFrame Original (Primeiras 5 linhas):")
display(df_original.head())
````

````python
# 2. Obter os nomes das colunas originais
original_column_names = df_original.columns.tolist()

# 3. Traduzir cada nome de coluna usando google.colab.ai
translated_column_names = []
print("\n### Traduzindo nomes das colunas...")
for col_name in original_column_names:
    prompt = f"Traduza o seguinte nome de coluna em inglês para português: '{col_name}'. Retorne apenas a tradução." # Adicionado instrução para retornar apenas a tradução
    translated_text = ai.generate_text(prompt)
    # Realiza uma limpeza simples para remover aspas ou formatação extra da resposta da IA
    cleaned_translation = translated_text.strip().replace("'", "").replace('"', '')
    translated_column_names.append(cleaned_translation)
    print(f"Original: '{col_name}' -> Traduzido: '{cleaned_translation}'")

# 4. Criar um novo DataFrame com os nomes das colunas traduzidos
df_translated = df_original.copy()
df_translated.columns = translated_column_names

print("\n### DataFrame com Colunas Traduzidas (Primeiras 5 linhas):")
display(df_translated.head())
````
